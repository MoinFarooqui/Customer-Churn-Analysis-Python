Converted O and 1 values of senior citizen to yes/no to make it easier to undertand 

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
    
    df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(
    x='Churn',
    data=df,
    palette=['#006400', '#8B0000']  # Dark Green, Dark Red
)

for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom'
    )

plt.title("Count of Customers by Churn")
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn': 'count'})

plt.pie(
    gb['Churn'],
    labels=gb.index,
    autopct='%1.2f%%',
    colors=['#006400', '#8B0000']
)

plt.title("Percentage of Churned Customers", fontsize=12)
plt.show()
```

From the given pie chart we can coclude that 26.54% of our customers have churned out
And explore the reason behind it 

```python
plt.figure(figsize=(4,4))

sns.countplot(
    x="gender",
    data=df,
    hue="Churn",
    palette=['#006400', '#8B0000']
)

plt.title("Gender Distribution by Churn")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
# Create a cross-tab and convert counts to percentages
ct = pd.crosstab(df['gender'], df['Churn'], normalize='index') * 100

# Plot 100% stacked bar chart
ax = ct.plot(
    kind='bar',
    stacked=True,
    figsize=(6,4),
    color=['#006400', '#8B0000']
)

# Add percentage labels manually
for i in range(len(ct)):
    cumulative = 0
    for j in range(len(ct.columns)):
        value = ct.iloc[i, j]
        if value > 0:
            ax.text(
                i,
                cumulative + value/2,
                f'{value:.1f}%',
                ha='center',
                va='center',
                color='white',
                fontsize=9,
                fontweight='bold'
            )
        cumulative += value

plt.title("Churn Percentage by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.ylim(0,100)

plt.show()
```

```python
# Create a cross-tab and convert counts to percentages
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot 100% stacked bar chart
ax = ct.plot(
    kind='bar',
    stacked=True,
    figsize=(5,4),
    color=['#006400', '#8B0000']
)

# Add percentage labels manually
for i in range(len(ct)):
    cumulative = 0
    for j in range(len(ct.columns)):
        value = ct.iloc[i, j]
        if value > 0:
            ax.text(
                i,
                cumulative + value / 2,
                f'{value:.1f}%',
                ha='center',
                va='center',
                color='white',
                fontsize=10,
                fontweight='bold'
            )
        cumulative += value

plt.title("Churn Percentage by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.ylim(0, 100)

plt.show()
```

```python
cat_cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

fig, axes = plt.subplots(6, 3, figsize=(18, 24))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(
        x=col,
        data=df,
        hue='Churn',
        palette=['#006400', '#8B0000'],
        ax=axes[i]
    )

    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

for j in range(len(cat_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Analysis of Categorical Features

The following count plots illustrate the distribution of customers across various categorical variables, with churn status represented by different colors. This analysis helps identify potential relationships between customer attributes, service usage, contract types, payment methods, and customer churn.

```python
plt.figure(figsize=(8,4))

ax = sns.countplot(
    x="PaymentMethod",
    data=df,
    hue="Churn",
    palette=['#006400', '#8B0000']
)

# Add count labels
for p in ax.patches:
    ax.annotate(
        str(int(p.get_height())),
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=9,
        fontweight='bold'
    )

plt.title("Customer Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=25)

plt.show()
```

Customer is likely to churn when he is using electronic check as a payment method
